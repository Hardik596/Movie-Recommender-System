```python
import ast # ast.literal.eval converts the stin to list
def gen_to_tags(a):
    l=[]
    for i in ast.literal_eval(a): 
        l.append(i['name'])
    print(l)
    return l
```

```python
gen_to_tags(df.iloc[4073].genres)
```

```python
#applying this to each of the row
df['genres']=df['genres'].apply(gen_to_tags)
```

```python
df.head()
```

```python
df['keywords'] = df['keywords'].apply(gen_to_tags)
```

```python
# we are taking the top 3 cast only
def top_cast(a):
    l=[]
    count = 0
    for i in ast.literal_eval(a):
        if len(l) <3:
            l.append(i['name'])
    count+=1
    
    return l
```

```python
df['cast']=df['cast'].apply(top_cast)
```

```python
df.iloc[0].crew
```

```python
def crew_director(obj):
    for i in ast.literal_eval(obj):
        if i['job'] == "Director" :
            return i['name']
```

```python
df['crew'] = df['crew'].apply(crew_director)
```

```python
len(df[df.crew.isnull()])
df.dropna(inplace=True)
```

```python
#making overview ready to become tag
df['overview']=df['overview'].apply(lambda x:x.split())
```

```python
#now we will remove the spaces with the names so that the name will have only single entity
df['genres'] =df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#now we will combine the keywords,overview,cast,crew into tags 
df['tags'] = df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
#now join it to make it a string so that we can use vectorization
def join(a):
    return " ".join(a)
df['tags']=df['tags'].apply(join)
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def preprocessor(a):
        review = a.lower()
        review = review.split()
        review = [lemmatizer.lemmatize(word) for word in review if len(word)>2 if word not in stopwords.words('english')]
        return review
```

```python
df['tags']= df['tags'].apply(preprocessor)
```

```python
df=df[['movie_id','title','tags']]
```

```python
#then do avgword2vec
#after that we can easily check the similar movies of it
```

```python
#training our word2vec model from scratch
import gensim
from gensim.models import word2vec,keyedvectors
model = gensim.models.Word2Vec(df['tags'])
```

```python
def avg_word2vec(doc):
    result = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    if result:
        return np.mean(result,axis=0)
    else:
        return np.zeros(model.vector_size)
```

```python
from tqdm import tqdm
x=[]
for i in tqdm(range(len(df['tags']))):
    x.append(avg_word2vec(df['tags'].iloc[i]))
```

```python
def recommendor_system(movie) :
    # index of the movie in df
    movie_index = df[df['title']==movie].index[0]
    distances = similarity[movie_index]
    #we will sort the distances and to get the correct index we use enumerate
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list :
        print(df.iloc[i[0]].title)
```
